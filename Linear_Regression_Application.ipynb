# **1. MỞ ĐẦU**

Các mô hình hồi quy mô tả mối quan hệ giữa các biến bằng cách khớp một đường thẳng với dữ liệu được quan
sát. Mô hình hồi quy tuyến tính sử dụng đường thẳng, trong khi mô hình hồi quy logistic và phi tuyến sử dụng
đường cong. Hồi quy cho phép bạn ước tính mức độ thay đổi của một biến phụ thuộc khi (các) biến độc lập thay
đổi. Mô hình hồi quy tuyến tính chia thành 2 dạng

## **1.1 KHÁI NIỆM MÔ HÌNH HỒI QUY TUYẾN TÍNH**

### **1.1.1 Mô hình hồi quy tuyến tính đơn biến**

Mô hình có công thức là:

$$
y = \beta_0 + \beta_1 X + \varepsilon
$$

Trong đó:

- y : Giá trị dự đoán của biến phụ thuộc (ký hiệu thường là \( y \)) cho một giá trị cụ thể của biến độc lập \( X \). Nói cách khác, \( y \) là giá trị mà mô hình hồi quy dự đoán cho biến phụ thuộc dựa trên một giá trị nhất định của biến độc lập.

- X : Biến độc lập (ký hiệu thường là \( X \)), là biến được cho là ảnh hưởng đến biến phụ thuộc \( y \).

- **\$ \beta_0 $**: Hệ số chặn (intercept), là giá trị dự đoán của \( y \) khi biến độc lập \( X = 0 \). Đây là điểm mà đường hồi quy cắt trục tung (trục \( y \)).

-  $ \beta_1 $\: Hệ số hồi quy (regression coefficient), thể hiện mức độ thay đổi dự kiến của biến phụ thuộc \( y \) khi biến độc lập \( x \) tăng thêm một đơn vị. \( $\beta_1 $\) cho biết hướng và độ dốc của đường hồi quy.

- $ \varepsilon $: Sai số, thể hiện mức độ sai lệch giữa giá trị dự đoán (ŷ) và giá trị thực tế của biến phụ thuộc. Giá trị \( $\varepsilon $\) càng nhỏ thì mô hình hồi quy càng chính xác.

### **1.1.2 Mô hình hồi quy tuyến tính bội**

Mô hình có công thức là

$$
y = \beta_0 + \beta_1 X_1 + + \beta_2 X_2+...+ \varepsilon
$$

Trong đó:

- y: Giá trị dự đoán của biến phụ thuộc (ký hiệu thường là y) cho một giá trị cụ thể của biến độc lập X. Nói cách khác, y là giá trị mà mô hình hồi quy dự đoán cho biến phụ thuộc dựa trên một giá trị nhất định của biến độc lập.

- X: Biến độc lập (ký hiệu thường là X), là biến được cho là ảnh hưởng đến biến phụ thuộc y.

- $\beta_0$: Hằng số hồi quy, hay còn được gọi là hệ số chặn. Đây là chỉ số nói lên giá trị của y sẽ là bao nhiêu nếu tất cả X cùng bằng 0. Nói cách khác, chỉ số này cho biết giá trị của y là bao nhiêu nếu không có các X. Khi biểu diễn trên đồ thị Oxy, $\beta_0$ là điểm trên trục Oy mà đường hồi quy cắt qua.

- $ \beta_1, \beta_2, ..., \beta_n $: Hệ số hồi quy, hay còn được gọi là hệ số góc. Chỉ số này cho biết về mức thay đổi của y gây ra bởi X tương ứng. Nói cách khác, chỉ số này nói lên có bao nhiêu đơn vị y sẽ thay đổi nếu X tăng hoặc giảm một đơn vị.

- $\epsilon$: Sai số, thể hiện mức độ sai lệch giữa giá trị dự đoán y và giá trị thực tế của biến phụ thuộc. Giá trị epsilon càng nhỏ thì mô hình hồi quy càng chính xác.

# **2. ỨNG DỤNG CỦA MÔ HÌNH HỒI QUY TUYẾN TÍNH**

## **2.1 ỨNG DỤNG CỦA MÔ HÌNH HỒI QUY TUYẾN TÍNH TRONG LĨNH VỰC KINH TẾ**

### **2.1.1 Vai trò của mô hình hồi quy tuyến tính trong lĩnh vực kinh tế**

Trong lĩnh vực kinh tế, mối tương quan giữa các biến là điều thường gặp. Tuy nhiên, xác định liệu mối tương
quan giữa hai hoặc nhiều biến có đại diện cho một mối quan hệ nhân quả hay không thì lại hiếm khi dễ dàng. Ví dụ:
* Các quốc gia giao thương nhiều hơn với thế giới thường có mức thu nhập cao hơn - nhưng liệu điều này có nghĩa là thương mại làm tăng mức thu nhập?
* Những người có trình độ học vấn cao hơn thường có thu nhập cao hơn, nhưng điều này có ngụ ý rằng giáo dục dẫn đến thu nhập cao hơn không?

Biết được câu trả lời chính xác cho những câu hỏi này là rất quan trọng. Nếu việc học thêm nhiều năm thực sự dẫn đến thu nhập cao hơn, thì các nhà hoạch định chính sách có thể giảm nghèo bằng cách cung cấp thêm ngân sách cho giáo dục. Nếu một năm học thêm dẫn đến mức tăng thu nhập 20.000 đô la một năm, thì lợi ích của việc chi tiêu cho giáo dục sẽ lớn hơn nhiều so với việc một năm học thêm chỉ giúp tăng thu nhập 2 đô la một năm.

Để giúp trả lời những câu hỏi kiểu này, các nhà kinh tế học sử dụng một công cụ thống kê được gọi là **phân tích hồi quy**. Hồi quy được sử dụng để lượng hóa mối quan hệ giữa một biến và các biến khác được cho là giải
thích cho nó; hồi quy cũng có thể xác định mức độ chặt chẽ và xác định rõ ràng của mối quan hệ đó. Ngày nay, việc
chạy hàng nghìn phân tích hồi quy đã trở nên phổ biến và dễ dàng - mặc dù trước đây không phải lúc nào cũng vậy
vì tuy những ý tưởng ban đầu về hồi quy có thể bắt nguồn từ thế kỷ 19, nhưng chính cuộc cách mạng công nghệ
của thế kỷ 20, biến máy tính để bàn trở thành công cụ thiết yếu, đã đẩy phân tích hồi quy lên một tầm cao mới.
Trong những năm 1950 và 1960, các nhà kinh tế học phải tính toán hồi quy bằng máy tính cơ học dùng điện. Thậm
chí cho đến tận năm 1970, việc nhận được kết quả của một phân tích hồi quy từ phòng thí nghiệm máy tính trung
tâm có thể mất đến 24 giờ - và đó là sau khi đã dành hàng giờ hoặc hàng ngày để đục lỗ trên thẻ tính. Một lỗi
đục lỗ (như từ khóa bị đánh vần sai hoặc giá trị dữ liệu không chính xác) có thể khiến toàn bộ nỗ lực trở nên vô nghĩa.Trên thực tế, rất khó để tìm thấy một nghiên cứu kinh tế thực nghiệm nào không có phân tích hồi quy. Các
lĩnh vực khác, bao gồm xã hội học, thống kê và tâm lý học, cũng rất tin cậy vào phân tích hồi quy


### **2.1.2 Những ứng dụng của mô hình hồi quy tuyến tính trong lĩnh vực kinh tế**

* **Dự đoán xu hướng thị trường**: Mô hình hồi quy có thể được sử dụng để dự đoán xu hướng thị trường trong
tương lai dựa trên các biến số kinh tế như GDP, lãi suất, tỷ lệ thất nghiệp, hay giá cả hàng hóa.
* **Dự đoán giá cổ phiếu**: Mô hình hồi quy cũng được sử dụng để dự đoán giá cổ phiếu dựa trên các biến số
như lợi nhuận, tỷ lệ thanh khoản, hay thông tin về doanh nghiệp.
* **Dự đoán doanh số bán hàng**: Doanh nghiệp có thể sử dụng mô hình hồi quy để dự đoán doanh số bán hàng
dựa trên các yếu tố như quảng cáo, giá cả, hoặc thị trường tiêu thụ. Điều này giúp họ tối ưu hoá chiến lược kinh
doanh và dự báo thành công cho tương lai.
* **Đánh giá tác động của chính sách kinh tế**: Mô hình hồi quy có thể giúp đánh giá tác động của các chính
sách kinh tế, chẳng hạn như chính sách tiền tệ hoặc chính sách thuế, đối với các biến số kinh tế khác bằng cách xây
dựng các mô hình mô phỏng các tác động của các chính sách khác nhau đối với các biến kinh tế quan trọng.
* **Phân tích rủi ro tín dụng**: Trong lĩnh vực ngân hàng, mô hình hồi quy cũng được sử dụng để phân
tích rủi ro tín dụng bằng cách dự đoán khả năng trả nợ của khách hàng dựa trên các biến số tài chính, các nhà
kinh tế sử dụng các mô hình để đánh giá rủi ro liên quan đến các khoản đầu tư, cho vay và các quyết định tài chính khác.
* **Lập mô hình kinh tế**: Bằng kết hợp các mô hình hồi quy vào các mô hình kinh tế vĩ mô và vi mô để mô
phỏng các hệ thống kinh tế phức tạp. Những ứng dụng trên chỉ là một số ví dụ cơ bản của việc sử dụng mô hình hồi
quy trong kinh tế, qua đó để có thể hiểu rõ hơn về những ứng dụng trên, dưới đây là một số ví dụ thực tiễn để
chúng ta tham khảo và hiểu rõ hơn về cách nó vận hành như thế nào:
* **Ứng dụng mô hình hồi quy không gian trong phân tích tăng trưởng doanh thu du lịch của các tỉnh/thành tại Việt Nam**: Mục đích của nghiên cứu là xác định và đo lường các yếu tố tác động đến doanh thu du lịch của các tỉnh/thành tại Việt Nam thông qua ứng dụng mô hình hồi quy không gian. Đề tài sử dụng
phương pháp nghiên cứu định tính kết hợp định lượng. Nghiên cứu định tính được thực hiện thông qua phỏng
vấn sâu cùng 09 chuyên gia. Nghiên cứu định lượng được thực hiện thông qua ứng dụng mô hình kinh tế lượng
không gian với bộ dữ liệu thứ cấp từ năm 2013 đến 2017 của các tỉnh/thành tại Việt Nam. Kết quả nghiên
cứu cho thấy: doanh thu du lịch của mỗi địa phương chịu tác động bởi: lượng khách, dân số và khối lượng luân
chuyển hành khách. Thêm vào đó, kết quả nghiên cứu là một bằng chứng thực nghiệm cho thấy có sự tương
quan giữa các tỉnh gần nhau trong tăng trưởng doanh thu du lịch, điều này có nghĩa là tăng trưởng du lịch
của địa phương này có tác động đến tăng trưởng doanh thu của các tỉnh lân cận. Đồng thời, lượng khách, dân
số và khối lượng luân chuyển của hành khách có sự tương tác về mặt không gian giữa các tỉnh/thành lân cận.
• Phương pháp hồi quy bội trong dự báo và ứng dụng vào dự báo doanh thu dịch vụ viễn thông
tại Thái Nguyên: Trong phần này, bài báo đề xuất giải pháp áp dụng phương pháp phân tích và dự báo
“hồi quy bội” vào dự báo doanh thu của dịch vụ viễn thông Thái Nguyên. Trên cơ sở nghiên cứu về các yếu
tố chủ yếu tác động đến doanh thu trong hoạt động kinh doanh của dịch vụ viễn thông Thái Nguyên, cụ thể đó là:

+ Dân số
+ Số thuê bao
+ Thu nhập

* **Do dự báo doanh thu dịch vụ viễn thông dựa vào 3 yếu tố:** Dân số, số thuê bao và thu nhập, mà 3 yếu
tố này có tính chất độc lập nhau trong quá trình nghiên cứu, nên phương pháp phân tích và dự báo “hồi quy
bội” được sử dụng ứng với 3 biến độc lập.
*  **Áp dụng mô hình toán kinh tế để phân tích các yếu tố ảnh hưởng và dự báo sản lượng điện trong tương lai:** Bằng cách nghiên cứu mối quan hệ tác động giữa nhu cầu tiêu thụ điện, tổng thu nhập thực
tế và giá điện dựa trên dữ liệu khảo sát của chính phủ về thu nhập thực tế (RIC) và giá điện trung bình ở Việt
Nam. Kết quả nghiên cứu chỉ ra rằng có mối liên quan rõ ràng khi thu nhập tăng lên, thì nhu cầu tiêu thụ
điện tăng lên. Giá điện cũng ảnh hưởng tiêu cực đến mức tiêu thụ điện, giá điện tăng thì tiêu thụ điện giảm

## **2.2 ỨNG DỤNG CỦA MÔ HÌNH HỒI QUY TUYẾN TÍNH TRONG LĨNH VỰC Y HỌC**

### **2.2.1 Vai trò của mô hình hồi quy tuyến tính trong lĩnh vực y học**

Hồi quy tuyến tính là kỹ thuật hồi quy cơ bản và được sử dụng phổ biến nhất
trong lĩnh vực nghiên cứu y học. Hồi quy tuyến tính được chia thành hai loại chính: hồi quy đơn biến và hồi quy bội.

* **Hồi quy tuyến tính đơn giản (Simple linear regression):**
Sử dụng khi có một biến phụ thuộc và một biến độc lập.

Ví dụ: Nghiên cứu câu hỏi "Khi tuổi tác tăng lên, nguy cơ mắc bệnh tiểu đường có tăng lên không, được đo bằng HbA1C hoặc lượng đường trong máu?".

Cả hai biến phải là biến liên tục (dữ liệu định lượng) và đường biểu diễn mối quan hệ giữa chúng là một đường thẳng (tuyến tính).

* **Hồi quy tuyến tính bội (Multiple linear regression):**
Sử dụng khi có một biến phụ thuộc liên tục và hai hoặc nhiều biến độc lập.

Ví dụ: Nghiên cứu câu hỏi "Nếu bệnh nhân mắc cả tiểu đường và tăng huyết áp, thì nguy cơ mắc bệnh động
mạch vành (CAD) của họ là bao nhiêu?".

Điều quan trọng là các biến độc lập có thể là biến định lượng hoặc biến định tính. Trong ví dụ trên, cả hai biến
độc lập đều có thể được biểu thị dưới dạng dữ liệu liên tục (giá trị huyết áp hoặc HbA1C) hoặc dữ liệu định tính (sự
có mặt hoặc không có bệnh tiểu đường theo tiêu chuẩn ADA 2016 hoặc tăng huyết áp theo tiêu chuẩn JNC VIII).

Cần có một mối quan hệ tuyến tính giữa biến phụ thuộc và các biến độc lập

### **2.2.2 Những ứng dụng của mô hình hồi quy tuyến tính trong lĩnh vực y học**

- Nghiên cứu nguy cơ và yếu tố bảo vệ: Một trong những ứng dụng chính của phân tích hồi quy trong y tế công
cộng là trong việc nghiên cứu về yếu tố nguy cơ và yếu tố bảo vệ. Chúng ta sử dụng phân tích hồi quy để đánh giá
mối liên hệ giữa các yếu tố này và các kết quả sức khỏe như mắc bệnh hoặc tử vong.

- Dự đoán kết quả lâm sàng: Hồi quy tuyến tính có thể được sử dụng để dự đoán kết quả lâm sàng như tình

trạng bệnh, phản ứng với điều trị, hoặc dự đoán nguy cơ mắc bệnh.
- Đánh giá nguy cơ: Bằng cách xem xét mối quan hệ giữa các yếu tố nguy cơ và kết quả y học, hồi quy tuyến
tính có thể giúp chúng ta hiểu được ảnh hưởng của các yếu tố này đối với sức khỏe.

- Nghiên cứu phác đồ điều trị: Trong các nghiên cứu lâm sàng, hồi quy tuyến tính có thể được sử dụng để phân
tích tác động của các phác đồ điều trị lên kết quả điểm cuối y học.

- Phân tích mối quan hệ giữa biến số lâm sàng: Hồi quy tuyến tính cũng có thể giúp chúng ta hiểu mối quan hệ
giữa các biến số lâm sàng, từ đó đưa ra các phát hiện quan trọng cho y học lâm sàng.

- So sánh các biện pháp đơn và nhiều mục về sức khỏe tự đánh giá chung (GSRH) để dự đoán tỷ lệ tử vong và
các sự kiện lâm sàng một số lượng lớn bệnh nhân kỳ cựu.

+ Mục tiêu : Các nhà nghiên cứu đã so sánh các biện pháp đơn và nhiều mục về sức khỏe tự đánh giá chung
(GSRH) để dự đoán tỷ lệ tử vong và các sự kiện lâm sàng một số lượng lớn bệnh nhân cựu chiến binh.

+ Nguồn dữ liệu/Môi trường nghiên cứu: Những nhà nghiên cứu đã phân tích dữ liệu đoàn hệ tiềm năng được

## **2.3 RỦI RO TIỀM ẨN CỦA PHÂN TÍCH HỒI QUY**

Mặc dù có nhiều lợi ích, phân tích hồi quy vẫn có những rủi ro nhất định và thường bị sử dụng sai cách. Dưới
đây là bốn khó khăn chính thường gặp:

**Biến số bị bỏ qua:**

Để xây dựng một mô hình hồi quy tốt, cần có một lý thuyết vững chắc để đề xuất các biến số có thể giải thích
cho biến phụ thuộc. Trong trường hợp hồi quy đơn giản với hai biến, cần phải nghĩ đến các yếu tố khác có thể ảnh
hưởng đến biến phụ thuộc.

Ví dụ, ngay cả khi đã tính đến IQ, mối quan hệ giữa trình độ học vấn và thu nhập có thể phản ánh một yếu tố
khác chưa được đưa vào mô hình. Điều này có nghĩa là các cá nhân trong mẫu lấy mẫu vẫn có thể khác nhau về
một số yếu tố "không quan sát được"giải thích cho thu nhập sau này của họ, có thể thông qua lựa chọn về học vấn.
Ví dụ, những người xuất thân từ gia đình giàu có thường có điều kiện học hành tốt hơn, nhưng sự giàu có của gia
đình cũng có thể tạo ra nhiều mối quan hệ hơn trên thị trường lao động, dẫn đến thu nhập cao hơn. Do đó, tài sản
của cha mẹ có thể là một biến số khác cần được đưa vào mô hình.

**Vấn đề nhân quả nghịch/nội sinh**

Trong phân tích hồi quy, nhân quả nghịch là một thách thức quan trọng cần lưu ý. Điều này xảy ra khi biến phụ
thuộc (biến được giải thích) cũng có thể ảnh hưởng đến các biến độc lập (biến dùng để giải thích). Điều này làm
phức tạp việc giải thích kết quả hồi quy vì hướng nhân quả thực sự giữa các biến trở nên không rõ ràng.

**Ví dụ:**

* Nghiên cứu mối quan hệ giữa trình độ học vấn và thu nhập. Phân tích hồi quy có thể cho thấy mối tương
quan giữa hai biến này, nghĩa là những người có trình độ học vấn cao hơn thường có thu nhập cao hơn.

* Tuy nhiên, có thể tồn tại vấn đề nhân quả nghịch.. Thu nhập cao hơn có thể cho phép mọi người đầu tư nhiều
hơn vào giáo dục của chính họ, từ đó lại làm tăng thu nhập của họ.

**Đo Lường Sai Sót:**

Các yếu tố có thể được đo lường không chính xác. Ví dụ, năng lực học tập rất khó đo lường và các bài
kiểm tra IQ cũng có những vấn đề rắc rối. Kết quả là, phân tích hồi quy sử dụng IQ có thể không chặt chẽ về
năng lực học tập, dẫn đến mối quan hệ tương quan không chính xác hoặc bị sai lệch giữa trình độ học vấn và thu nhập.

**Tầm Nhìn Quá Hẹp:**
Hệ số hồi quy chỉ cung cấp thông tin về việc các thay đổi nhỏ (không phải thay đổi lớn) trong một biến số liên
quan đến các thay đổi trong biến số khác như thế nào. Hệ số hồi quy sẽ cho thấy một thay đổi nhỏ về trình độ học
vấn có thể ảnh hưởng đến thu nhập như thế nào nhưng không cho phép nhà nghiên cứu khái quát hóa về tác động
của những thay đổi lớn. Nếu tất cả mọi người cùng học đại học cùng một lúc, thì một người mới tốt nghiệp đại học
có lẽ sẽ không kiếm được nhiều tiền hơn vì tổng cung sinh viên đại học sẽ tăng lên đáng kể.

Biểu đồ Q-Q phần dư

```r
secondBodyFatModel<-lm(bodyfat$BodyFat~bodyfat$Age+bodyfat$Weight+bodyfat$Height
+bodyfat$Ankle+bodyfat$Forearm+bodyfat$Wrist)
```

cao nên sẽ không xảy ra hiện tượng đa cộng tuyến.

### **3.2.4 Nhận xét mô hình**

Xét hai mô hình bodyFatModel và secondBodyFatModel để biết liệu mô hình bodyFatModel hay secondBodyFatModel
tốt hơn. Một cách để trả lời câu hỏi này là chạy phân tích phương sai (ANOVA) cho hai mô hình. Phân tích ANOVA
kiểm định giả thuyết không (H0) - cho rằng các biến đã loại bỏ trước đó không có ý nghĩa, đối với giả thuyết đối
(H1) - cho rằng các biến đó có ý nghĩa. Cách diễn giải kết quả ANOVA:

*   Nếu mô hình mới là cải tiến so với mô hình ban đầu: Thì chúng ta không bác bỏ được H0. Điều này cho thấy
việc loại bỏ các biến là hợp lý và mô hình mới không bị ảnh hưởng đáng kể bởi các biến đó.

*  Nếu mô hình mới không tốt hơn mô hình cũ: Thì điều này có nghĩa là các biến đã loại bỏ có ý nghĩa. Do đó,
chúng ta bác bỏ H0. Có thể cần xem xét việc đưa các biến này trở lại mô hình hoặc phân tích thêm để hiểu
tại sao mô hình mới lại kém hiệu quả hơn.

Kết quả của hàm anova

```r
anova(bodyFatModel,secondBodyFatModel)
```

Dựa vào kết quả ANOVA, chúng ta thấy giá trị p (2.2e-16) rất nhỏ (bé hơn 0.05). Do đó, bác bỏ giả thuyết
không (H0). Điều này có nghĩa là mô hình thứ hai không phải là sự cải thiện so với mô hình thứ nhất.
Ngoài ra, việc dựa vào R-squared hiệu chỉnh cũng là cách để tìm ra mô hình phù hợp hơn. Dưới đây là kết quả
R-squared hiệu chỉnh của mô hình ban đầu và mô hình thứ hai:

Mô hình ban đầu có adjusted R-squared (R-squared điều chỉnh) là 0.7353, cao hơn so với adjusted R-squared của
mô hình thứ hai (0.5628).Mô hình ban đầu cũng thỏa mãn ba giả định nên điều này cho thấy mô hình ban đầu với
tất cả các biến dự báo là tốt hơn so với mô hình thứ hai.
